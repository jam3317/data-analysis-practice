# 7.2 args, kwargs



```python
def print_args(*args):
    print(args)
    for x in args:
        print(x)

print_args(100, True, 'leehojun')
```

```python
def print_kwargs(a, **kwargs):
    print(a)
    print(kwargs)
    for i in kwargs:
        print(i)

print_kwargs(100, name='leehojun', age='10')
```

```python
# 숫자를 입력 받고, 해당 값의 약수를 구하는 함수를 작성하시오.

def divisor(num):

  divisors = []

  for i in range(1, num+1):
    if num % i == 0:
      divisors.append(i)

  return divisors


num = int(input('숫자를 입력해주세요 : '))

divisor(num)
```

```python
# 함수는 리스트와 알파벳을 입력받습니다. 입력받은 리스트에서 알파벳이 몇 개 있는지 출력하는 함수를 만들어주세요.

def solution(l, a):
    return l.count(a)

solution(['a', 'b', 'c', 'a', 'a'], 'a') # 정답 : 3
```

```python
# 함수는 2차원 리스트를 입력받습니다. 아래와 같이 2차원 리스트를 두번째 값을 기준으로 내림차순 정렬하는 함수를 만들어주세요.

#sort() 는 원본을 변형시키고, sorted()는 원본을 건드리지않는다.

def 함수(x):
  return x[1]

def solution(l):
  return sorted(l, key=함수, reverse=True)

solution([[40, 5], [20, 3], [30, 4], [10, 1]])
# 정답 : [[10, 5], [30, 4], [20, 3], [40, 1]]
```

```python
# 2차원 리스트가 주어졌을 때 각 값의 차를 저장한 리스트를 출력해주세요.

def solution(l):
  answer = []
  for i in l:
    answer.append(i[0] - i[1])
  return answer

solution([[10, 5], [20, 3], [30, 4], [40, 1]])
# 정답 : [[5], [17], [26], [39]]
```

```python
#주어진 문자열의 합을 구하는 코드를 작성하세요.

def solution(s):

    return sum([int(i) for i in s])

solution('11123')
# 정답 : 8
```
