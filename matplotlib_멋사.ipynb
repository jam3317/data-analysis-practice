## 1. Matplotlib

- 데이터를 Chart, Plot으로 시각화해주는 라이브러리이다.
- 2003년 0.1 출시, 19년이 된 패키지(DataFrame이 나오기 이전부터 있던 시각화 패키지)
- 가장 많이 사용되는 시각화 패키지이다
- 기본 구성 : 그림(figure, 이미지 출력 단위), 축(axes)

- ! : 콘솔에서 사용 가능한 명령어를 사용가능하게 해줌
- fig.savefig() : figure에 있는 이미지를 저장

### **Basic Attributes**

- alpha : 투명도
- logy : Y축 Log scaling
- kind : line, bar, barh, kde
- subplots : 여러개의 plot 그리기
- legend : subplot의 범례 지정
- xlim, ylim : X축과 Y축의 경계(axis로 한번에 그릴 수 있음)
- grid : 그리드 표현(True, False)
- title : 그래프의 제목 지정
- linewidth : 라인 넓이
- color : 색
- linestyle : 실선, 점선, 1점 쇄선 등
- marker : 마커 지정
- markerfacecolor : 마커 색
- markersize : 마커 크기
- 그 외 Attribute : https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html

## 3. Scatter Plot

- 변수들의 상관관계, 밀집 위치를 나타낸다.
- 두 변수의 상관관계를 직교 좌표계의 평면에 점으로 표현하는 그래프
- 양의 상관관계, 음의 상관관계
- 군집 : 점들의 모임

```python
plt.scatter(20,20, s=100, c='r', alpha=0.5)

plt.show()
```

```python
plt.scatter(20,20, s=1000, c='g', alpha=0.5)

plt.show()
```

```python
plt.scatter([10,20],[10,20], s=[100,300], c=['g','r'], alpha=0.5)

plt.show()
```

```python
x = np.linspace(0, 10, 21)
y = x ** 2

plt.scatter(x,y, s=x*100, c='r', alpha=0.5)

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
color = np.random.rand(50)

size = 1000*np.random.rand(50)

plt.scatter(x,y, s=size, c=color, alpha=0.5)
plt.show()
```

## 4. **subplot**

- 여러개의 그래프를 하나의 그림에 나타내도록 함
- `plt.subplot(row, col, index)`의 형태임(콤마 없이 사용 가능)

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(x,y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(221)
plt.plot(x,y, 'r')

plt.subplot(222)
plt.plot(x,y, 'g')

plt.subplot(223)
plt.plot(x,y, 'b')

plt.subplot(224)
plt.plot(x,y, 'y')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(211)
plt.plot(x,y, 'r')
plt.xticks(visible=False)

plt.subplot(212)
plt.plot(x,y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

x_ = np.linspace(0, 20 , 41)
y_ = x_ ** 2

ax1 = plt.subplot(211)
plt.plot(x,y, 'r')
plt.xticks(visible=False)

ax2 = plt.subplot(212, sharex=ax1)
plt.plot(x_, y_, 'g')

plt.show()
```
