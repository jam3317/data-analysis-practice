```python
import ole

with ole.open('data.hwp') as f:
    data = f.get_stream('PrvText').read().decode('utf-16le')
    print(data)
```

```python
# >< 기준으로 나누고, 띄어쓰기 없애기

#data.split('><')[1].replace(' ','') == '성명'
data.split('><')[2]
```

```python
splitdata = data.split('><')
```

```python
for data in splitdata:
  #print(data)
  if data.replace(' ','') == '성명':
    이름인덱스 = splitdata.index(data) + 1

splitdata[이름인덱스]
```
