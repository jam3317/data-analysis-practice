### **Basic Attributes**

- alpha : 투명도
- logy : Y축 Log scaling
- kind : line, bar, barh, kde
- subplots : 여러개의 plot 그리기
- legend : subplot의 범례 지정
- xlim, ylim : X축과 Y축의 경계(axis로 한번에 그릴 수 있음)
- grid : 그리드 표현(True, False)
- title : 그래프의 제목 지정
- linewidth : 라인 넓이
- color : 색
- linestyle : 실선, 점선, 1점 쇄선 등
- marker : 마커 지정
- markerfacecolor : 마커 색
- markersize : 마커 크기
- 그 외 Attribute : https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html

## 4. **subplot**

- 여러개의 그래프를 하나의 그림에 나타내도록 함
- `plt.subplot(row, col, index)`의 형태임(콤마 없이 사용 가능)

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(x,y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(221)
plt.plot(x,y, 'r')

plt.subplot(222)
plt.plot(x,y, 'g')

plt.subplot(223)
plt.plot(x,y, 'b')

plt.subplot(224)
plt.plot(x,y, 'y')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

plt.subplot(211)
plt.plot(x,y, 'r')
plt.xticks(visible=False)

plt.subplot(212)
plt.plot(x,y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10 , 21)
y = x ** 2

x_ = np.linspace(0, 20 , 41)
y_ = x_ ** 2

ax1 = plt.subplot(211)
plt.plot(x,y, 'r')
plt.xticks(visible=False)

ax2 = plt.subplot(212, sharex=ax1)
plt.plot(x_, y_, 'g')

plt.show()
```

### **선 색상과 스타일**

- 공식문서 : https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html
- 공식문서 : https://matplotlib.org/3.2.1/api/pyplot_summary.html

---

- plt.plot(x, y, color='red') -> red, green, blue 등 색상 이름
- plt.plot(x, y, color='r') -> r, g, b, y, m(자홍), k(검정) 등 색상 이름
- plt.plot(x, y, color='0.2') -> 회색조(0-1사이 값)
- plt.plot(x, y, color='#ff0000') -> 16진수 색상 값

---

- plt.plot(x, y, linestyle='solid') -> 실선('-')
- plt.plot(x, y, linestyle='dashed') -> 파선('--')
- plt.plot(x, y, linestyle='dashdot') -> 1점 쇄선('-.')
- plt.plot(x, y, linestyle='dotted') -> 점선(':')

---

- plt.plot(x, y, '--r') -> 빨간색 파선

---

- '.' point marker
- 'o' circle marker(plt.plot(x, y_, '-o'))
- '^' triangle_up marker
- 's' square marker
- 'p' pentagon marker
- '*' star marker
- '+' plus marker
- 'x' x marker

## 5. Bar Chart

- 범주형 자료의 분포를 파악한다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
label = ['one', 'two', 'three']
value = [10,20,30]

plt.bar(label, value, color='red')
plt.show()
```

```python
label = ['one', 'two', 'three']
value = [10,20,30]
color = ['red', 'green', 'blue']

plt.bar(label, value, color=color, width=0.5)
plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=20)
plt.ylabel('helloy', fontsize=20)
plt.show()
```

```python
label = ['one', 'two', 'three']
value = [10,20,30]
color = ['red', 'green', 'blue']

plt.barh(label, value, color=color, height=0.5) #눕히기
plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=20)
plt.ylabel('helloy', fontsize=20)
plt.show()
```

```python
label = ['one', 'two', 'three']
value = [-10,-20,30] #음수도 가능, 항아리 모양의 그래프
color = ['red', 'green', 'blue']

plt.barh(label, value, color=color, height=0.5) #눕히기
plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=20)
plt.ylabel('helloy', fontsize=20)
plt.show()
```

```python
#누적 바 그래프

plt.bar(['one', 'two', 'three'],[10,20,30], color='g')
plt.bar(['one', 'two', 'three'],[20,40,60], color='b',
         alpha=0.5, bottom=[10,20,30])

plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)
plt.legend(['chart1', 'chart2'])

plt.show()
```

```python
plt.bar(['one', 'two', 'three'],[10,20,30], color='g')
plt.bar(['one', 'two', 'three'],[20,40,60], color='b',bottom=[10,20,30])

plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)
plt.legend(['chart1', 'chart2'])

plt.show()
```

## 6. histogram

- 도수분포표를 그래프로 나타낸 방법이다.
- 막대그래프는 계급 즉 가로를 생각하지 않고 세로의 높이로만 나타낸다. (출처 : 위키백과)
- 연속형 자료를 계급으로 나누어  계급별 도수를 막대로 나타낸다.

```python
x = [np.random.randint(1,7) for i in range(100000)] #1~6 까지 랜덤으로 100000번 출력, 주사위

#print(x)
```

```python
plt.hist(x, bins=11) #구간 개수

#plt.xlim([0.5, 7.5]) #x좌표 범위 늘리기

plt.show()
```

```python
#키 조사하기

height = [173,182,163,175,192,189,190,200,176]
print(sorted(height))

plt.hist(height)

plt.show()
```

```python
#키 조사하기

height = [173,182,163,175,192,189,190,200,176]

plt.hist(height, bins=10, alpha=0.5)
plt.hist(height, bins=20, alpha=0.5)

plt.show()
```

```python
height = [173,182,163,175,192,189,190,200,176]

plt.hist(height, cumulative=True, label='one')
plt.hist(height, cumulative=False, label='two')

plt.legend(loc='upper left')
plt.show()
```

## 7. Pie Chart

```python
labels = ['one', 'two', 'three']
size = [100,20,10]

plt.pie(size, labels=labels)

plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100,20,10]

plt.pie(size, labels=labels, shadow=True) #그림자 주기

plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100,20,10]

plt.pie(size, labels=labels, shadow=True, autopct='%1.3f%%') #소수점 둘째자리까지

plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100,20,10]
explode = (0,0.1,0.3) # 떼기

plt.pie(size, labels=labels, shadow=True, explode=explode, autopct='%1.3f%%') #소수점 둘째자리까지

plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [50,30,20]
explode = (0,0,0) # 떼기

plt.pie(size, labels=labels, shadow=True, explode=explode, autopct='%1.3f%%', startangle=90) #90도에서 시작 (12시 부터 시작)

plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [50,30,20]
explode = (0,0,0) # 떼기

plt.pie(size, labels=labels, shadow=True, explode=explode, autopct='%1.3f%%', startangle=90, counterclock=False) #90도에서 시작 (12시 부터 시작), 및 12시부터 순서대로 출력

plt.show()
```
