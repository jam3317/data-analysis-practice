 4. 복소수형

## 1.2 NumPy 차원

- Scalar : 0차원 하나의 데이터 값으로만 존재하는 것

- Tensor : 3차원 이상의 다차원 배열

````python
# 0차원 (Scalar)

import numpy as np

a = np.array(1)

print(a)
print(a.shape)
print(a.ndim) #n dimention
````

````python
# 2차원 (Matrix)

import numpy as np

a = np.array([[1,2,3],[4,5,6]])

print(a)
print(a.shape)
print(a.ndim) #n dimention
````

````python
# 3차원 (Tensor)

import numpy as np

a = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])

print(a)
print(a.shape)
print(a.ndim) #n dimention
````

### 1.3.1 배열의 연결과 분할

1. 연결
    - hstack : 열 추가
    - vstack : 행 추가
    - concatenate ([a, b], axis = ) : axis=0 → 행 추가, axis=1 →  열추가
2. 분할
    - vsplit : 수직축으로 분할, 행 분할
    - hsplit : 수평축으로 분할, 열 분할

````python
a = [1,2,3]
b = [4,5,6]

a = np.array(a)
a = np.array(b)

a+b
````

````python
#concatenate([a,b]) -> 두 배열 더하기

np.concatenate([a,b])
````

````python
a = np.arange(10).reshape(2,5)
b = np.arange(10, 20).reshape(2,5)

print(a)
print(b)
````

````python
#concatenate 로 2차원 배열 더하기

print(np.concatenate([a,b]))

print()
print()

print(np.concatenate([[a],[b]])) #각각 묶어서 더하기

````

````python
# 축 설정해서 더하기 (axis)

print(np.concatenate([a,b], axis=0))
print()
print(np.concatenate([a,b], axis=1))
````

````python
# vstack, hstack

print(np.vstack([a,b]))
print()
print(np.hstack([a,b]))
````

## 1.7 random

- seed : 시작 숫자를 정해 정해진 난수 알고리즘으로 난수 생성
- shuffle : 데이터의 순서를 바꿔줌
- choice : 데이터에서 일부를 무작위로 선택
- rand : 0부터 1까지 균일 분포로 난수 생성
- randn : 가우시안 표준 정규 분포로 난수 생성

````python
import random as r
````

````python
r.randint(1,10) # 1과 10을 포함한 난수 생성
````

````python
r.random() # 0 <= 부동 소수점 숫자 <= 1 생성
````

````python
r.uniform(1,10) # 1 <= 부동 소수점 숫자 <= 10 생성
````

````python
# 1,3,5,7,9 에서의 난수 생성

r.randrange(1,10,2)
````

````python
l = [i for i in range(20)]
l
````

````python
# 배열의 인덱스를 마구잡이로 섞음

r.shuffle(l)
l
````

````python
# 아무 숫자나 골라서 출력

r.choice(l)
````

````python
# 무작위로 샘플 원한다면

r.sample(l, 5)
````

````python
#난수를 고정

r.seed(7)
r.randint(1,10)
````

````python
np.random.randint(1,10)
````

````python
np.random.normal(1,10,size=(10)) # 정규 분포로 임의 샘플
np.random.uniform(1,10,size=(3,4))
````

````python
np.random.seed(1)
np.random.randint(1,10)
````
