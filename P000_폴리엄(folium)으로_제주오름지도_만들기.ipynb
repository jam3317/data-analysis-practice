- 다른 라이브러리보다 오래되어 안정적인 라이브러리

```python
m = folium.Map(
    location = [33.3684955195788, 126.52918183373025],
    zoom_start = 11
)

tooltip = 'hello!'

folium.Marker(
    [33.3684955195788, 126.52918183373025],
    popup = "<strong>hello</strong>~~!!<img width='100' height='180' src = 'https://i.pinimg.com/736x/c3/27/07/c327075970f3809a5523ce774e22ff94.jpg'",
    tooltip = tooltip
).add_to(m)

m
```

```python
m = folium.Map(
    location = [33.3684955195788, 126.52918183373025],
    zoom_start = 10
)

folium.Marker(
    [33.3684955195788, 126.52918183373025],
    popup = "<strong>한라산</strong><img width='100' height='180'",
    tooltip = '한라산!',
    icon = folium.Icon(color='red', icon='info-sign')
).add_to(m)

folium.Marker(
    [33.5070772, 126.4934311],
    popup = "<strong>제주국제공항</strong><img width='150' height='200'",
    tooltip = '제주국제공행',
    icon = folium.Icon(color='##ff0000', icon='bookmark')
).add_to(m)

folium.Marker(
    [33.5083269, 126.5410764],
    popup = "<strong>바울랩</strong><img width='150' height='200'",
    tooltip = '바울랩'
).add_to(m)
m
```

```python
#전체 오름 출력

for i in range(오름.shape[0]):
  print(오름.iloc[i]['오름명'], 오름.iloc[i]['위도'], 오름.iloc[i]['경도'])
```

```python
m = folium.Map(
    location = [33.3684955195788, 126.52918183373025],
    zoom_start = 10
)

folium.Marker(
    [33.5083269, 126.5410764],
    popup = "<strong>바울랩</strong>",
    tooltip = '바울랩'
).add_to(m)

#전체 오름 마커

for i in range(오름.shape[0]):

  folium.Marker(
    [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
    popup = f'<strong>{오름.iloc[i]["오름명"]}</strong>',
    tooltip = 오름.iloc[i]
  ).add_to(m)


m
```

```python
m = folium.Map(
    location = [33.3684955195788, 126.52918183373025],
    zoom_start = 10
)

tooltip = 'Click!!'

for i in range(오름.shape[0]):
    folium.Marker(
        [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
        주차장 : {오름.iloc[i]["주차장"]}<br>\
        화장실 : {오름.iloc[i]["화장실"]}<br>\
        </div>',
        tooltip = tooltip
    ).add_to(m)

m
```

```python
m.save('index.html')
```
